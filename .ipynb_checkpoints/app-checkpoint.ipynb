# 一、Android Studio Setup

1、首先下载Android studio安装包

下载路径：[Android studio]https://developer.android.google.cn/studio

创建AndroidTool目录，再创建三个子目录，分别为AndroidStudio，AndroidSDK和AndroidProject

2、进入Android Studio安装程序，按照如下操作进行：

1）点击Next



2）选择第二选项，点击Next



3）保存在自己的安装路径下，这里我用的是子目录AndroidStudio



4）选择Android Studio，点击Install



5）等待安装，成功后运行Android Studio，创建第一个项目



6）第一次运行项目会下载gradle，等候下载完成后编译成功，即创建项目完成



# 二、Jupyter Notebook 安装

1、先更新到最新版的pip3



2、以管理员身份运行“命令提示符”，安装Notebook：pip install notebook



3、运行Notebook：jupyter notebook



4、Jupyter Notebook安装成功界面



# 三、Notebook启动之后，默认的Files列出了用户文件夹的项目，那么如何更改默认加载的目录呢？

1）首先查找到默认用户文件夹项目的文件：jupyter notebook --generate-config



2）找到该文件，寻找notebook_dir，修改到自己想要的文件夹



3）到jupyter notebook中查看结果
